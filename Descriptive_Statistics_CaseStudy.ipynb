```python
#Interquartile (  Q3-Q1)
IQR=q3-q1
```

```python
IQR
```

```python
#upper fence
upperfence=q3+1.5*IQR
print(upperfence)
```

```python
lowerfence=q1-1.5*IQR
```

```python
print(lowerfence)
```

```python
#any value lesser than lowerfence i.e 27712 and greater than upperfence i.e 47336 is known as outlier.
```

```python
#corelation
#Correaltion will alwayes be found for the whole dataset and  not for the specific column
#so here we are going to find correlation for the whole df and not the columns i.e. population i.e.population in thousands(2017)
```
