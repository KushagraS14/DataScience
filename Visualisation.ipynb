```python
#line plot
height=[121.9,124.5,129.5,134.5,139.7,147.3,152.4,157.5,162.6]
weight=[19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2]

plt.plot(height,weight,marker='o',color='green',linestyle='--')
for i in range(len(height)):
  text=''
  text=str(height[i])+','+str(weight[i])
  plt.annotate(text,(height[i],weight[i]))

plt.xlabel("Height in centimeters")
plt.title('Height vs Weight chart')
plt.ylabel("weight in kgs")
plt.legend('Height vs Weight')
plt.show()
```

```python
#Pie chart
#budget  for projects 60M,marketing=10M,hr=5M,finance =10M
budget=[60,10,5,25]
myexplode=[0,0,0.1,0]
names=['projects','maketing','hr','finance']
plt.pie(budget,labels=names,explode=myexplode,shadow=True,autopct='%.1f' ,startangle=90)
plt.title('Project Pie chart')
plt.show()

```

```python
#creating arrow

#Quiver(X_pos,Y_pos,X_dir1,y_dir2)
import numpy as np
import matplotlib.pyplot as plt
X_pos1=0
Y_pos1=0
X_dir1=1
Y_dir1=1

X_pos2=0
Y_pos2=0
X_dir2=-1.5
Y_dir2=-1.5
u=np.cos(X_dir2)*Y_dir2
v=np.sin(Y_pos2)*Y_dir2
fig,axes=plt.subplots(figsize=(12,7))
axes.quiver(X_pos1,Y_pos1,X_dir1,Y_dir1)
axes.quiver(X_pos2,Y_pos2,u,v)
axes.set_title('The Quiver plot with one arrow')
#show plot
plt.show()




```
