```python
data=pd.read_csv('/content/cleaned_df.csv',index_col=0)
```

```python
data.head()
```

```python
#age: persons 'age in years
#sex perons se(1=male and 0: female)
#cp:chest pain type
#value 0 aymtomatic
#Value1 :Atypical angina
#Value2 Non anginal pain
#Value3 A typical angina
#trestbps:The persons resting blood presuure(mm Hg on admission to Hospital)
#chol:The person's cholestrol measurement in mg/dl
#fbs:THe persons fasting blood sugar(>120 mgdl 1=True 0=False)
#restecg resting electrocardiographic results
#-Value 0 showing probable or definite leftventicular hypertrophy by Estes' criteria
#value 1 :Normal
#value2:having St-T wave abnormality(T-wave inversions and ST evaluations or depressions of >0.05Mv)
#thalach:The person's maximumheart rate achieved
#exang Exercise induced angina(1=yes 0=no)
#old peak:ST DEPRESSION INDUCED BY EXERCISE RELTIVE TO REST('ST' RELATES TO positions on the ECG PLOT .SEE MORE)
#slope The slope ofthe Peak exercise  ST segment --0 downsloping 1:flat 2:upsloping
```

```python
data.shape
```

```python
X=data.drop('target',axis=1)
```

```python
X.head(3)
```

```python
y=data['target']
```

```python
y.head(4)
```

```python
#the datasets are in different range so we have to stadardise the dataset between same range then our machine learning algorithms works better
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
#dummy classifier like random classifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,f1_score,roc_auc_score,roc_curve
```

```python
log_reg=LogisticRegression()
log_reg.fit(X_train,y_train)
```

```python
y_pred_log=log_reg.predict(X_test)
```

```python
#accuracy calculation
lr_acc_score=accuracy_score(y_test,y_pred_log)
lr_f1_score=f1_score(y_test,y_pred_log)            # how accurate our model is for all classes TP ,FP,TN,FN
lr_conf_matrix=confusion_matrix(y_test,y_pred_log)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt
tree=DecisionTreeClassifier()
tree.fit(X_train,y_train)
plt.figure(figsize=(30,10))
plot_tree(tree,filled=True,feature_names=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang',	'oldpeak','slope','	ca','thal','target']) # filled=True is given to give the heatmap kind of analysis Darker colors have more weightage ,the lighter colors have leser weightage.overall in the tree which nodes end up having more importance
plt.show() #Decision tree can be visualised in two ways
```
